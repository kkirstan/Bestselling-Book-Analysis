```python
# Dependencies
from bs4 import BeautifulSoup
import requests
from requests import get
import pandas as pd
import numpy as np
```

```python
from time import sleep
from random import randint
```

```python
#!!!
# URL of page to be scraped
url = 'https://www.barnesandnoble.com/b/books/_/N-1fZ29Z8q8?Nrpp=20&page='
```

```python
#!!!
# Retrieve page with the requests module
response = requests.get(url)
```

```python
#!!!
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
#!!!
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
titles = []
authors = []
pb_dates = []
book_urls = []
```

```python
pages = (1,2,3,4,5,6)
```

```python
#!!
for page in pages:
    request = requests.get("https://www.barnesandnoble.com/b/books/_/N-1fZ29Z8q8?Nrpp=20&page=" + str(page))
    soup = BeautifulSoup(request.text, 'html.parser')
    results = soup.find_all('div', class_="col-lg-9 product-info-listView")
    sleep(randint(2,10))
    
    for result in results:
        try:
            # Identify and return title of listing
            title = result.find('a', class_="").text
            titles.append(title)
        
            #Identify the author
            author = result.find('div', class_="product-shelf-author contributors").text
            authors.append(author)
        
            # Identify and return publication date of listing
            publication_date = result.find('span', class_="text--small publ-date").text
            pb_dates.append(publication_date)
            
            # Identify and return link to listing
            link = result.a['href']
            book_urls.append(link)
            
            # Print results only if title, price, and link are available
            if (title and author and publication_date and link):
                print('-------------')
                print(title)
                print(author)
                print(publication_date)
                print("https://barnesandnoble.com" + link)
        except AttributeError as e:
            print(e)
```

```python
books_df = pd.DataFrame({
'Title': titles,
'Author': authors,
'Publication Date': pb_dates,
})
```

```python
books_df.to_csv('bn.csv')
```
